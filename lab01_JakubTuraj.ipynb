```python
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
  return np.maximum(0, x)

a = np.array([-3, -1, 0, 1, 3, 5])
print(f"ReLU array: {relu(a)}")

x_values = np.linspace(-5, 5, 100)
plt.plot(x_values, relu(x_values))
plt.ylabel("ReLU values")
plt.xlabel("x values")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')
stats = iris.groupby('species')['petal_length'].agg(['min', 'max', 'mean', 'std'])
print(stats)
print(f"\nMost consistent species: {stats['std'].idxmin()}")
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

features = iris.columns[:4]
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
for i, feat in enumerate(features):
    ax = axes[i//2, i%2]
    sns.histplot(iris[feat], ax=ax, kde=True)
    ax.set_title(f"Dist of {feat}")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

X = iris.drop(columns='species').values
X_manual = (X - X.mean(axis=0)) / X.std(axis=0)
#print(X_manual)

scaler = StandardScaler()
X_sklearn = scaler.fit_transform(X)
#print(X_sklearn)
print(f"Manual matches Sklearn: {np.allclose(X_manual, X_sklearn)}")
```

```python
from matplotlib.lines import lineStyles
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

W = np.random.randn(5, 4)
x = np.random.randn(4)
res = sigmoid(np.dot(W, x))

print(f"Result shape: {res.shape}")
print(f"Output: {res}")

def plot_stats(data):
  mu, sigma = data.mean(), data.std()
  plt.hist(data, bins=30, alpha=0.7, color='springgreen', edgecolor='white')
  plt.axvline(mu, color='red', linestyle='solid', linewidth=2, label='Mean')
  plt.axvline(mu - sigma, color='red', linestyle='dashed', label='Std Dev')
  plt.axvline(mu + sigma, color='red', linestyle='dashed', label='Std Dev')
  plt.legend()
  plt.show()

data = np.random.normal(0, 1, 1000)
plot_stats(data)
```
